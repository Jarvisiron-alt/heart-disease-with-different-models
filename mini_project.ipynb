```python
import numpy as np 
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("D:\Downloads\heart.csv")
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
for i in df.columns:
    for each in df[i].values:
        
        if each > 1 or each < -1:
            df[i] = (df[i] - np.min(df[i]))/(np.max(df[i]) - np.min(df[i]))
        else:
            pass
y = df["target"]
X = df.drop(["target"],axis = 1)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 4)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, confusion_matrix, roc_curve, classification_report
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from collections import Counter
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
log_reg=LogisticRegression(random_state=0,max_iter=900)
log_reg.fit(X_train,y_train)
y_pred=log_reg.predict(X_test)
log_reg.score(X_test,y_test)
```

```python
params = {'C': np.logspace(-3, 3, 7), 'penalty': ['l1', 'l2']}
lr_model = LogisticRegression(random_state = 0)
lr_cv = GridSearchCV(lr_model,params,cv = 5).fit(X_train,y_train)
lr_cv.best_params_
dir(lr_model)
```

```python
log_reg2 = LogisticRegression(C = 1,random_state = 0,penalty='l2')
log_reg2.fit(X_train,y_train)
```

```python
log_reg2_accuracy = log_reg2.score(X_test, y_test)
log_reg2_accuracy
```

```python
knn = KNeighborsClassifier(n_neighbors = 2).fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
params = {"n_neighbors": range(1,50)}
knn_model = KNeighborsClassifier()
knn_cv = GridSearchCV(knn_model, params, cv = 10).fit(X_train,y_train)
knn_cv.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors = 8).fit(X_train,y_train)
knn_accuracy=knn.score(X_test,y_test)
knn_accuracy
```

```python
tree = DecisionTreeClassifier(random_state = 0).fit(X_train,y_train)
tree.score(X_test,y_test)
```

```python
params = {"max_depth": range(1,10),
            "min_samples_split" : list(range(2,50))}
tree_model = DecisionTreeClassifier(random_state = 0)
tree_cv = GridSearchCV(tree_model, params, cv = 10).fit(X_train,y_train)
tree_cv.best_params_
```

```python
tree_tuned = DecisionTreeClassifier(max_depth = 7, min_samples_split = 37,random_state = 42).fit(X_train,y_train)
tree_tuned_accuracy=tree_tuned.score(X_test,y_test)
tree_tuned_accuracy
```

```python
svm = SVC(random_state = 42).fit(X_train,y_train)
svm.score(X_test,y_test)
```

```python
params = {"C": [0.0001, 0.001, 0.1, 1, 5, 10 ,50 ,100],
             "gamma": [0.0001, 0.001, 0.1, 1, 5, 10 ,50 ,100]}
svm_model = SVC(random_state = 42)
svm_cv = GridSearchCV(svm_model,params,cv = 10).fit(X_train,y_train)
svm_cv.best_params_
```

```python
svm_tuned = SVC(C = 100, gamma = 0.0001, random_state = 42).fit(X_train,y_train)
svm_tuned_accuracy = svm_tuned.score(X_test,y_test)
svm_tuned_accuracy 
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pred_list = [log_reg2,knn,tree_tuned,svm_tuned]

for i in pred_list:
    print("Score : ",i.score(X_test,y_test))
    y_pred = i.predict(X_test)
    sns.heatmap(confusion_matrix(y_test,y_pred),annot = True)
    plt.xlabel("Y_pred")
    plt.ylabel("Y_test")
    plt.title(i)
    plt.show()
```
